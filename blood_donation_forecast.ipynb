```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("transfusion.data")

# Split features and target variable
X = data.drop(columns=['whether he/she donated blood in March 2007'])
y = data['whether he/she donated blood in March 2007']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=77, random_state=8)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
predictions = rf_classifier.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# Display classification report
print(classification_report(y_test, predictions))

# Plot the graph
plt.figure(figsize=(8, 6))
plt.bar(['Accuracy'], [accuracy], color=['blue'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.show()

# Plot the distribution of the target variable
plt.figure(figsize=(6, 4))
data['whether he/she donated blood in March 2007'].value_counts().plot(kind='bar', color=['blue', 'red'])
plt.title('Blood Donation Distribution')
plt.xlabel('Donated Blood in March 2007')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.show()
```
